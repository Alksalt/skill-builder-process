# reshaping, stacking, squeezing and unsqueezing
* Reshaping an input tensor to a defined shape
* View - Return a view of an input tensor of certain sha[pe but keep the same memory as the original tensor
* Stacking - combine multiple tensors on top of each other (Ovstack) or side by side (hstack)
* Squeeze - removes all 1 dim from a tensor
* Unsqueeze - add a 1 dim to a target tensor
* Permute - return a view of the input with dim permuted(swapped) in a certain way

```python
# reshaping, stacking, squeezing and unsqueezing

x = torch.arange(1., 11.)
x, x.shape
```

```python
# Add extra dim
x_reshaped = x.reshape(2,5 )
x_reshaped, x_reshaped.shape
```

```python
#CHange the view
z = x.view(1, 10)
z, z.shape
```

```python
# Stack tensors
x_stacked = torch.stack([x,x,x,x], dim=0)
x_stacked
```

```python
# .squeeze and usqueeze
x = torch.zeros(2, 1, 2, 1, 2)
print(x.size())
y = torch.squeeze(x)
print(y.size())
y = torch.squeeze(x, 0)
print(y.size())
y = torch.squeeze(x, 1)
print(y.size())
y = torch.squeeze(x, (1, 2, 3))
print(y.size())
```

```python
### torch.permute is a PyTorch function used to reorder
x = torch.randn(10, 3, 64, 64) # Example tensor with shape (batch_size, channels, height, width)
print("Original shape:", x.shape)

# Permute dimensions to (batch_size, height, width, channels)
y = torch.permute(x, (0, 2, 3, 1))
print("Permuted shape:", y.shape)
```

#Reproducbility(trying to take random out of random)


```python
t_A = torch.rand(3,4)
t_B = torch.rand(3,4)
print(t_A == t_B)
```

```python
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
t_C = torch.rand(3,4)

torch.manual_seed(RANDOM_SEED)
t_D = torch.rand(3,4)

print(t_C)
print(t_D)
print(t_C == t_D)
```
